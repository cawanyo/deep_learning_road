```python
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

# Loading the dataset 

```python
boston_housing = keras.datasets.boston_housing
(X_train, y_train), (X_test, y_test) = boston_housing.load_data()

#Normalize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(f"Train data shape: {X_train.shape}, Test data shape: {X_test.shape}")
```

# Build the model

```python
model = keras.Sequential()
model.add(keras.layers.Dense(16, activation='relu'))
model.add(keras.layers.Dense(8, activation='relu'))
model.add(keras.layers.Dense(1))

#optimization parameter
model.compile(optimizer='Adam', loss='mse', metrics=['mse'])
```

```python
#Training the model
model.fit(X_train, y_train, epochs=1000)
```

```python
model.evaluate(X_test, y_test)
```

# Analyze
We can see that our model error on the training data is way less than the one on the test set.<br> So we can see that our model overfit or that we have <b> variance problem</b> <br> <br>
So we will try some regularization technique to overcome that
<ul>
    <li> L2 Regularization </li>
    <li> Dropout </li>
<ul>

```python
model = keras.Sequential()
model.add(keras.layers.Dense(16, activation='relu', kernel_regularizer=keras.regularizers.l2(l=0.1)))
model.add(keras.layers.Dense(8, activation='relu', kernel_regularizer=keras.regularizers.l2(l=0.1)))
model.add(keras.layers.Dense(1))

#optimization parameter
model.compile(optimizer='Adam', loss='mse', metrics=['mse'])
```

```python
#Training the model
model.fit(X_train, y_train, epochs=1000)
```

```python
model.evaluate(X_test, y_test)
```

We can see that we already improve the test error

```python
model = keras.Sequential()
model.add(keras.layers.Dense(16, activation='relu'))
model.add(keras.layers.Dropout(0.1))
model.add(keras.layers.Dense(8, activation='relu'))
model.add(keras.layers.Dense(1))

#optimization parameter
model.compile(optimizer='Adam', loss='mse', metrics=['mse'])
```

```python
#Training the model
model.fit(X_train, y_train, epochs=1000)
```

```python
model.evaluate(X_test, y_test)
```

In this exemple we don't see the improvement bring by the dropout <br>but in most of the case dropout is a powerfull things

# Impact of initializer
Now let's see how the initializer can impact the performance of a model<br>
We can check the initializer list on the keras <a href="https://keras.io/api/layers/initializers/"> website </a> <br>
Every initializer work good in a special scenario <br>
In the future post we will see that on differents type of project

```python
model = keras.Sequential()
model.add(keras.layers.Dense(16, activation='relu', kernel_initializer= keras.initializers.HeNormal))
model.add(keras.layers.Dense(8, activation='relu', kernel_initializer= keras.initializers.HeNormal))
model.add(keras.layers.Dense(1))

#optimization parameter
model.compile(optimizer='Adam', loss='mse', metrics=['mse'])
```

```python
#Training the model
model.fit(X_train, y_train, epochs=1000)
```
